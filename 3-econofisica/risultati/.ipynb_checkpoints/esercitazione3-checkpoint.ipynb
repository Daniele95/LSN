Non possiamo mai seguire i dettagli del movimento del granello (particella browniana), che subisce una svolta ad ogni collisione molecolare, circa $10^{12}$ volte al secondo in un liquido ordinario. Ciò che osserviamo in termini di spostamenti sono delle fluttuazioni residue attorno a un valore medio zero e hanno poco collegamento diretto con il percorso dettagliato effettivo del granello; per i nostri sensi (spinti al limite massimo sotto forma della miglior telecamera cinematografica che scatta a $10^5$ al secondo), i dettagli del percorso sono estremamente fini. Il percorso può essere paragonato in modo approssimativo al grafico di una funzione continua senza derivata. 

Osserviamo un granello browniano al tempo $0$ e al tempo $t$. Durante l'intervallo di tempo $(0, t)$ ha subito uno spostamento $\Delta s$, la cui proiezione sull'asse $x$ è $\Delta x$. Lo stesso granello viene osservato in tempi successivi $2t, 3t, ...$, e $\Delta x$ viene determinato per ogni intervallo. Questi valori vengono elevati al quadrato e ne viene calcolato il valore medio, che chiameremo $\Delta x^2$. Sottolineiamo che questi spostamenti che osserviamo non sono in alcun modo il percorso dettagliato del granello, né $\Delta x/t$ è la sua velocità. Ad esempio, in un intervallo di tempo $t = 1~\text{secondo}$, la particella compie milioni di collisioni e ciò che vediamo sono semplicemente le sue posizioni iniziali e finali, che collegiamo con una linea retta, mentre il vero percorso è un confuso zigzag di segmenti lineari. Quello che osserviamo è già un "percorso" notevolmente semplificato, un esempio del quale è dato nella Figura. 

<img src="rw2.png" width=60%>

Se si immagina che ciascuno degli elementi lineari in questa figura sia composto da milioni di linee rette, si comincia ad avvicinarsi al percorso effettivo. È impossibile analizzare questo complicato movimento in tutti i suoi dettagli e dobbiamo quindi accontentarci di osservare, in un determinato intervallo di tempo $t$, la corrispondente grandezza di $\Delta x^2$, che è solo vagamente correlata al vero percorso. (Coffey, The Langevin Equation, 2005, World Scientific)

Se una particella soggetta a una forza di rumore bianco $\eta(t)$, con
\begin{eqnarray}
\langle\eta(t)\eta(t')\rangle=\sigma^2\delta(t-t'),
\end{eqnarray}
l'effetto complessivo di questa forza casuale sul moto della particella è la somma nel tempo di tutti gli urti ricevuti ovvero
$$W(t)=\int_0^t\eta(t')dt'$$

è possibile differenziare questa quantità secondo le regole del calcolo stocastico, considerando che mediamente una particella diffonde come $$\langle x^2\rangle=t$$ pertanto $$dW^2=dt$$


Se consideriamo che la particella è anche sogggetta a una forza di deriva deterministica, possiamo scrivere l'equazione differenziale del moto
\begin{eqnarray}
dx(t)=a(x,t)~dt+b(x,t)~dW(t)
\end{eqnarray}
dove $$x=x(t)$$

## Formula di Itô

Posso calcolare le derivate di una funzione $f(x(t),t)$ in questo modo:
\begin{eqnarray}
df(x,t)=\frac{\partial f}{\partial t}dt+\frac{\partial f}{\partial x}dx+
\frac{1}{2}\frac{\partial^2f}{\partial t^2}dt^2+
\frac{1}{2}\frac{\partial^2f}{\partial x^2}dx^2+
\frac{\partial^2f}{\partial t\partial x}dt~dx
\end{eqnarray}
Dal momento che \begin{eqnarray}dW^2=dt\end{eqnarray}
e che \begin{eqnarray}dt^m=0~\text{per}~m\geq 2,~~dW^m=0~\text{per}~m\geq 3,
~~ dtdW=0\end{eqnarray}
si ha
\begin{eqnarray}
df&=&\frac{\partial f}{\partial t}dt+\frac{\partial f}{\partial x} (a~dt+b~dW)
+\frac{1}{2}\frac{\partial^2 f}{\partial x^2}(a~dt+b~dW)^2
+\frac{\partial^2f}{\partial t\partial x}dt~(a~dt+b~dW)\\
&=&\frac{\partial f}{\partial t}dt+
a~\frac{\partial f}{\partial x}~dt+
b~\frac{\partial f}{\partial x}~dW+
\frac{1}{2}~b^2~\frac{\partial^2 f}{\partial x^2}~dW^2\\
&=&\left(\frac{\partial f}{\partial t}+
a~\frac{\partial f}{\partial x}+
\frac{b^2}{2}\frac{\partial^2 f}{\partial x^2}\right)~dt+
b~\frac{\partial f}{\partial x}~dW
\end{eqnarray}
che è la formula di Itô per il differenziale di una funzione $f$. 


Ad esempio, il differenziale della funzione $$y=\ln x$$ è $$dy=\frac{dx}{x}-\frac{dx^2}{2x^2}$$
poiché, dal momento che $dx$ contiene un incremento $dt$ e un incremento $dW\sim\sqrt{dt}$, lo sviluppo di Taylor al primo ordine deve contenere anche il termine quadratico, e si ha
$$
dy=a~dt+b~dW-\frac{1}{2 x^2}b^2x^2 dt
$$
\begin{eqnarray}
\text{d~ln}~x(t)&=&\left(a-\frac{b^2}{2}\right)dt+b~dW(t)
\end{eqnarray}
integrando,
\begin{eqnarray}\ln x=\left(a-\frac{b^2}{2}\right)~t+b~W(t)
\end{eqnarray}
esponenziando,
$$x(t)=e^{\left(a-\frac{b^2}{2}\right)~t+b~W(t)}$$
$$\langle x(t)\rangle=e^{\left(a-\frac{b^2}{2}\right)~t}\langle e^{b~W(t)}\rangle$$

questo si può calcolare tenendo presente che
$$d\langle e^{b~W(t)}\rangle=b~e^{b~W(t)}~dW+\frac{b^2}{2}~e^{b~W(t)}~dt$$
$$\frac{d\langle e^{b~W(t)}\rangle}{dt}=\frac{b^2}{2}\langle e^{b~W(t)}\rangle$$
e pertanto
$$\langle e^{b~W(t)}\rangle=e^{b^2t/2}$$ pertanto
$$\langle x(t)\rangle=e^a$$
il primo momento dipende soltanto dalla forza deterministica di deriva.

## Moto Browniano geometrico

Il moto Browniano geometrico è un processo di Itô
\begin{eqnarray}
ds=a(s,t)~dt+b(s,t)~dW(t)
\end{eqnarray}
nel quale
\begin{eqnarray}
a(s,t)~\sim s,~~b(s,t)~\sim s
\end{eqnarray}
Trattandosi di azioni, ci interessa il guadagno in proporzione al capitale iniziale $dS/S$, per questo possiamo considerare che la nostra variabile rilevante è $\text{ln}~S$. Si ha
\begin{eqnarray}
\text{d~ln}~S(t)&=&\left(\mu-\frac{\sigma^2}{2}\right)dt+\sigma~dW(t)
\end{eqnarray}
Pertanto la variabile casuale $d~ln~S$ ha distribuzione gaussiana con media
\begin{eqnarray}
\langle d~\ln~S(t)\rangle=\left(\mu-\frac{\sigma^2}{2}\right)~dt
\end{eqnarray}
e varianza
\begin{eqnarray}
\langle( d~\ln~S(t))^2\rangle-\langle d~\ln~S(t)\rangle^2=\sigma^2~dt
\end{eqnarray}
pertanto, se $S$ è un moto browniano geometrico con media $\mu$ e varianza $\sigma^2$, e ha valore iniziale $S_0$, allora
\begin{eqnarray}
S(t)=S_0~e^{\left(\mu-\frac{\sigma^2}{2}\right)~t+\sigma~W(t)}
\end{eqnarray}

## Modello di Black-Scholes

**Contratti nei mercati idealizzati**

Un investitore prudente desidera evitare il rischio e proteggersi da eventuali perdite. Come può farlo?
Un derivato è un prodotto finanziario il cui valore "deriva" dal prezzo di un asset sottostante. Matematicamente, se l'asset sottostante è caratterizzato dal suo prezzo, un derivato è semplicemente una funzione di $S(t)$.
Come primo esempio, un *forward* è un contratto tra due parti in cui un partner accetta di comprare un asset sottostante ("posizione lunga") in un certo momento specificato $T$ in futuro, noto come data di consegna, per un prezzo prescritto $K$, noto come prezzo di consegna, e l'altro partner si impegna a vendere l'asset sottostante al tempo $T$ per il prezzo $K$ ("posizione corta").
La motivazione di base per firmare un contratto, come un forward, è quella di eliminare il rischio di mercato dell'evoluzione imprevedibile dei prezzi.

**Opzioni**
Le opzioni sono contratti in cui solo una delle parti assume l'obbligo, mentre l'altra ottiene un diritto. Ci sono molti tipi di opzioni. Le più semplici sono le opzioni europee. Un'opzione europea è un contratto tra due parti in cui il venditore dell'opzione, noto come il *writer*, concede all'acquirente dell'opzione, noto come il titolare, il diritto di acquistare (opzione *call*) dal writer o di vendere (opzione *put*) a lui un sottostante (con un prezzo spot attuale $ S(t)$) per un prezzo prescritto $K$, chiamato prezzo di esercizio, alla data di scadenza $T$ in futuro.

La proprietà chiave di un'opzione è che solo il writer ha un obbligo. Deve vendere o comprare l'asset sottostante per il prezzo di esercizio al tempo $T$. Il titolare sfrutterà il suo diritto solo se ottiene un profitto, cioè se $ S(T) > K$ per un'opzione call. Altrimenti, può acquistare l'asset sottostante a un prezzo più basso, $S(T) < K$, sul mercato. Il profitto è $\max[0, S(T) - K]$ per una call, e da $\max[0, K - S(T)]$ per una put.
Naturalmente, il writer non assume questa obbligazione finanziaria senza richiedere una compensazione.

La domanda centrale, dunque, è: **quanto dovrebbe costare un'opzione?**
La teoria di Black-Scholes fornisce una risposta a questa domanda per le opzioni europee. L'analisi di Black-Scholes si basa su alcune ipotesi:

- <b>Accesso all'informazione</b>: Gli attori del mercato hanno accesso istantaneo e completo a tutte le informazioni rilevanti per il trading. Questo consente loro di prendere decisioni informate in modo rapido e accurato.

- <b>Liquidità</b>: Il mercato è altamente liquido, il che significa che gli investitori possono facilmente comprare o vendere prodotti finanziari in qualsiasi momento senza influenzare significativamente il prezzo di mercato. La liquidità fornisce agli investitori la sicurezza che possano sempre convertire i loro asset in denaro quando necessario, aumentando l'attrattiva del mercato e riducendo il rischio di investimento.

- <b>Basso attrito di mercato</b>: L'attrito di mercato, che comprende tutti i costi associati al trading, è minimo. Ciò significa che i costi di transazione sono trascurabili rispetto al volume complessivo delle transazioni. Un basso attrito di mercato favorisce una maggiore partecipazione degli investitori e contribuisce a mantenere i mercati efficienti.

Questi principi lavorano insieme per creare un ambiente in cui i prezzi dei prodotti finanziari riflettono accuratamente tutte le informazioni disponibili e dove gli investitori possono agire con fiducia sapendo che il mercato è efficiente, liquido e caratterizzato da costi di transazione minimi.

La teoria di Black-Scholes assume che l'evoluzione temporale del prezzo dell'asset sia stocastica e segua un moto browniano geometrico (GBM) con tasso di interesse privo di rischio $r$ costante e volatilità $\sigma$ costante.

Soluzione analitica di Black-Scholes per il prezzo al tempo t < T:

- Prezzo dell'opzione call europea:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- Prezzo dell'opzione put europea:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

dove $S$ è il prezzo dell'asset sottostante,
\begin{eqnarray}d_1 &=& \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right],~~~~~~
d_2 &=& d_1 - \sigma\sqrt{T-t}
\end{eqnarray}
e
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} $$

```python

# data
S0 = 100.
K = 100.
T = 1.
r = 0.1
sigma = 0.25
t=0

def N(x):
    return 0.5 * (1. + erf(x / sqrt(2.)))

def black_scholes(S0, K, T, r, sigma):
    d1 = 1./(sigma * sqrt(T)) * (log(S0 / K) + (r + (sigma**2) / 2.) * T)
    d2 = d1 - sigma * sqrt(T)
    C = S0 * N(d1) - K * exp(-r * T) * N(d2)
    P = S0 *(N(d1) - 1.) - K * exp(-r * T) * (N(d2)-1.)
    return C, P

if __name__ == "__main__":
    call, put = black_scholes(S0, K, T, r, sigma)
    print(" == BLACK-SCHOLES ==")
    print("call: ", call)
    print("put: ", put)
```

## Limiti dell'ipotesi di mercato efficiente

Nel 2003, 5 anni prima della grande recessione del 2008, Robert Lucas, vincitore del Premio Nobel per l'Economia nel 1995, dichiarò: <i>la macroeconomia ha avuto successo: il suo problema centrale di prevenzione della depressione è stato risolto, per tutti gli scopi pratici</i>. Eppure, ironicamente, il 19 ottobre 1987, appena poche settimane dopo la conferenza di Santa Fe, l'indice Dow Jones ha registrato il suo peggior giorno di sempre, crollando del 22,6% in un solo giorno. Un esempio lampante di un evento raro, totalmente assente dalla teoria classica dei mercati suppostamente gaussiani. Inoltre, questo crollo non sembra essere legato a nessuna notizia economica che possa giustificarne l'entità.

Di fronte a un evento del genere senza una causa apparente, il fisico pensa immediatamente a una discontinuità endogena, generata da cicli di feedback interni al complesso sistema che è un mercato finanziario, con i suoi migliaia di attori che si influenzano a vicenda... ma questa interpretazione era molto lontana dal dogma dominante all'epoca (e anche adesso), quello degli agenti razionali, la cui infinita saggezza permetterebbe ai mercati di mostrare sempre il prezzo "vero", quello che riflette il cosiddetto valore "fondamentale", e che varia solo se nuove informazioni lo giustificano. Questa è la teoria dei mercati efficienti e stabili, che non si lasciano trasportare spontaneamente, non cedono a movimenti di panico ingiustificati, ma ostinatamente forniscono informazioni sul valore del mondo. <i>Davvero credete a questo?</i> è stata la reazione netta e poco diplomatica di Phil Anderson quando questa teoria gli fu presentata a Santa Fe nel 1987.

La prima sorpresa è che, nel mondo di Black-Scholes, il premio assicurativo non dipende dal valore del rendimento medio atteso del mercato azionario nell'anno successivo; che il mercato azionario salga o scenda, il prezzo delle opzioni rimane invariato; pertanto, non è necessario fare alcuna ipotesi su questo rendimento medio per stabilire il prezzo dell'assicurazione contro la diminuzione dei prezzi delle azioni!

La seconda, forse ancora più inquietante sorpresa, è che l'assicuratore può seguire una strategia nota come </i>copertura</i>, che gli permette di non correre affatto rischi. Tuttavia, il modello assume esplicitamente che i prezzi di mercato siano imprevedibili. Come può, quindi, il rischio scomparire completamente?

La risposta risiede nella natura molto particolare delle ipotesi di Black-Scholes-Merton, che assumono che i movimenti dei prezzi siano descritti da un moto Browniano continuo nel tempo. Il moto Browniano può essere costruito come la sovrapposizione di un numero infinito di passi infinitesimamente piccoli. In un tale modello, il prezzo evolve continuamente, senza salti; il caso è 'benigno', l'incertezza può essere domata e, in un certo senso, scompare.

Ma è un modello singolare, fragile: non appena si tiene conto del fatto che le statistiche delle variazioni dei prezzi hanno code lunghe, che si verificano salti di prezzo di tutte le dimensioni nei mercati finanziari<a name="cite_ref-1"></a>[<sup>[1]</sup>](#cite_note-1), da quelli insignificanti a crolli di un secolo come quello del 1987, allora il rischio che Black e Scholes avevano nascosto sotto il tappeto riappare, ed è infatti lontano dall'essere trascurabile. Inoltre, questo rischio è asimmetrico: le potenziali perdite dell'assicuratore sono molto maggiori di quelle dell'assicurato. Se non fosse così, l'esistenza stessa dei contratti assicurativi di questo tipo sarebbe priva di senso, poiché l'assicurazione, per definizione, è un trasferimento di rischio. In un certo senso, il modello contraddice proprio l'esistenza dei contratti che cerca di teorizzare.

Ma le cose diventano ancora peggiori. L'uso indiscriminato del modello Black-Scholes-Merton crea un loop di feedback destabilizzante, che ha considerevolmente amplificato il crollo del 1987 che ho menzionato in precedenza. Quindi il paradosso è completo: l'uso di un modello in cui i crolli non esistono può effettivamente scatenarne l'esistenza! Il modello stesso diventa un fattore di rischio sistemico.

<a name="cite_note-1"></a><small>1. [^](#cite_ref-1)  Terremoti di magnitudo molto diverse coesistono, alcuni impercettibili e altri distruttivi. La scala Richter è logaritmica: +1 su questa scala corrisponde a un terremoto 30 volte più potente.</small>

Nel caso del modello di Black-Scholes-Merton, un ingrediente mancante ma essenziale è il feedback della strategia di copertura sui prezzi. Il prezzo di un asset finanziario non esiste di per sé, come vorrebbe farci credere la visione platonica dei mercati efficienti (e come hanno creduto a lungo gli ingegneri finanziari, e a volte ancora credono). Il prezzo è il risultato delle azioni stesse degli acquirenti e dei venditori. Le transazioni hanno un impatto sui prezzi, e questo impatto può innescare un loop di feedback e portare all'instabilità. (<a href="https://iopscience.iop.org/article/10.1088/2632-072X/ad104a">Bouchaud 2023, J. Phys. Complex. 4 041001</a>)
