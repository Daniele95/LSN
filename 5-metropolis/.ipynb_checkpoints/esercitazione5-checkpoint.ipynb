# Esercitazione 5: calcolo di orbitali atomici con l'algoritmo di Metropolis


Occorre calcolare i seguenti integrali: valor medio $\langle r\rangle$ sullo stato fondamentale

\begin{eqnarray}
\Psi_{100}=\frac{1}{\sqrt{\pi}}e^{-r}
\end{eqnarray}

\begin{eqnarray}
\langle 100|r|100\rangle&=&
\frac{1}{\pi}\int_\mathbb{R} r~e^{-2r} d^3x\\
&=&4\int_{-\infty}^\infty r^3~e^{-2r}~dr\\
&=&2.5
\end{eqnarray}

e sullo stato eccitato con momento angolare 1,


\begin{eqnarray}
\Psi_{210}=\frac{1}{\sqrt{32\pi}}re^{-r/2}\cos\theta
\end{eqnarray}


\begin{eqnarray}
\langle 210|r|210\rangle&=&
\frac{1}{32\pi} \int_\mathbb{R}r~r^2e^{-r}\cos^2\theta~d^3x\\
&=&\frac{1}{16}\int_{-\infty}^\infty r^5~e^{-r}~dr\int_{-\pi}^\pi cos^2\theta~d\theta\\
&=&5
\end{eqnarray}

Gli integrali si svolgono analiticamente e danno questi risultati. Possiamo confrontarli con i risultati che ricaviamo eseguendo il calcolo tramite l'algoritmo di Metropolis.

# Algoritmo di Metropolis

L'algoritmo di campionamento Metropolis è una tecnica avanzata descritta per la prima volta in un articolo di Metropolis, Rosenbluth, Rosenbluth, Teller e Teller. Il metodo è correlato alle tecniche di rifiuto poiché comporta esplicitamente la proposta di un valore provvisorio che può essere respinto e poiché la normalizzazione della funzione campionata è irrilevante, non è necessario conoscerla. Il metodo è anche collegato ai processi stocastici di Markov poiché il campionamento di una particolare densità di probabilità si ottiene costruendo un processo di Markov correlato e simulandolo. L'algoritmo di Metropolis è di grande semplicità e potenza; può essere utilizzato per campionare essenzialmente qualsiasi funzione di densità indipendentemente dalla complessità analitica in qualsiasi numero di dimensioni. 

Gli svantaggi sono che il campionamento è corretto solo asintoticamente e che le variabili successive prodotte sono correlate, spesso molto fortemente. Ciò significa che l'valutazione di integrali di solito produce correlazioni positive nei valori dell'integranda con conseguente aumento della varianza per un numero fisso di passaggi rispetto a campioni indipendenti. Il metodo è stato motivato da un'analogia con il comportamento dei sistemi nella meccanica statistica che si avvicinano a un equilibrio le cui proprietà statistiche sono indipendenti dalla cinetica del sistema. Con cinetica intendiamo una matrice di trasferimento stocastica $K(x|y)$ che governa l'evoluzione del sistema; non è altro che una distribuzione di probabilità condizionata ed è anche chiamata nucleo stocastico o probabilità di transizione. Nel trattare un sistema fisico, si presume di solito che $K(x|y)$ sia noto, e ci si prefigge di trovare $p(x)$. L'algoritmo di Metropolis inverte questo approccio: si ha il compito di trovare una cinetica (dinamica stocastica) conveniente e corretta che equilibrerà il sistema in modo che la data $p(x)$ risulti essere la probabilità di osservare il sistema vicino a $x$. Questo si rivela estremamente facile grazie al dispositivo elegante suggerito da Metropolis. Le transizioni sono proposte da, diciamo, $y$ a $x$ utilizzando essenzialmente "qualsiasi" distribuzione $T(x|y)$. Quindi, confrontando $p(x)$ con $p(y)$ e tenendo conto anche di $T$, il sistema viene spostato in $x$ (mossa "accettata") o riportato in $y$ (mossa "respinta").

<img src="metropolis.png" width=35%>

1. Si sceglie un punto di partenza $\vec{x}$.
2. Partendo da $\vec{x}$ si fa un salto casuale a $\vec{x}'$, con $\Delta\vec{x}=\vec{x}-\vec{x}'$ estratto da una distribuzione uniforme o gaussiana con un certo passo.
3. Si accetta il passo con probabilità data da $A(\vec{x}'|\vec{x})=\min\bigg[1,\frac{p(\vec{x}')}{p(\vec{x})}\bigg]$. In questo modo si accettano tutti i passi verso le regioni a maggiore probabilità, mentre quelli verso zone a probabilità minore sono accettati in modo casuale in base alla loro stessa probabilità.

Una buona regola empirica usata per coniugare una veloce convergenza a una correlazione non elevata vuole un tasso di accettazione $\sim$ 50%. Per ottenere tale percentuale è possibile regolare la larghezza della distribuzione che determina la grandezza dei salti.
Prima di equilibrare bisogna scegliere la larghezza del passo $\delta$. In base alle varie situazioni sono state scelti i seguenti parametri, i quali hanno dato tassi di accettazione compresi tra il 45% e il 55% circa.

| $\delta$ | Fondamentale | Eccitato |
|----------|--------------|----------|
| Uniforme | 1.4          | 2.5      |
| Gauss    | 0.8          | 1.7      |

La funzione d'onda è centrata nell'origine e simmetrica rispetto all'origine. Scegliamo però di mettere alla prova l'algoritmo di Metropolis, facendolo partire da una posizione iniziale abbastanza lontana dall'origine, e vediamo entro quanti passi si porta abbastanza vicino all'origine. Quando l'esploratore resta sempre attorno all'origine possiamo considerare la simulazione equilibrata, e iniziare da lì le misure.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


fig = plt.figure(figsize=(10,8))

def plot(plot,stato,coda):
    X, Y, Z = np.loadtxt('risultati/posizione'+stato+'Uniforme.txt',usecols=(0,1,2),unpack=True)
    if not(coda):
        X=X[:400]
        Y=Y[:400]
        Z=Z[:400]
    else:
        X=X[400:]
        Y=Y[400:]
        Z=Z[400:]
    ax = fig.add_subplot(1,2,plot, projection='3d')
    ax.scatter(X, Y, Z, c=Z, marker='.')
    ax.view_init(elev=20, azim=20)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('Stato '+stato,fontweight='bold',fontsize=16)
    bound=20
    ax.set_xlim(-bound,bound)
    ax.set_ylim(-bound,bound)
    ax.set_zlim(-bound,bound)

plot(1,'Fondamentale',False)
    
plot(2,'Eccitato',False)

plt.show()
```

Mostriamo ora il grafico della distanza media dell'esploratore dall'origine, in funzione del numero di passi (nel caso di passo uniforme).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

N = 400
x = np.arange(N)+1

plt.figure(figsize =(14,8))
plt.suptitle("Equilibrazione", fontsize=20)
plt.subplot(2,1,1)

dist = np.loadtxt("risultati/posizioneFondamentaleUniforme.txt",max_rows=400, unpack='true')
norms = np.linalg.norm(dist, axis=0)
plt.plot(x, norms, label = "Uniforme")
dist = np.loadtxt("risultati/posizioneFondamentaleGaussiano.txt",max_rows=400,  unpack='true')
norms = np.linalg.norm(dist, axis=0)
plt.plot(x, norms, label = "Gauss")
plt.ylabel('$r/a_0$', fontsize=13)
plt.xlabel('Passo', fontsize=13)
plt.grid(True)
plt.title('Distanza - Stato Fondamentale', fontsize=15, fontweight = "bold")
plt.legend()


plt.subplot(2,1,2)
dist = np.loadtxt("risultati/posizioneEccitatoUniforme.txt",max_rows=400, unpack='true')
norms = np.linalg.norm(dist, axis=0)
plt.plot(x, norms, label = "Uniforme")
dist = np.loadtxt("risultati/posizioneEccitatoGaussiano.txt", max_rows=400, unpack='true')
norms = np.linalg.norm(dist, axis=0)
plt.plot(x, norms, label = "Gauss")
plt.ylabel('$r/a_0$', fontsize=13)
plt.xlabel('Passo', fontsize=13)
plt.grid(True)
plt.title('Distanza - Stato Eccitato', fontsize=15, fontweight = "bold")
plt.legend()

plt.subplots_adjust(hspace=0.5)

plt.show()
```

Come si può vedere, entro 400 passi la simulazione può ben dirsi equilibrata, dato che si attesta in buona approssimazione attorno a un valore medio rispettivamente di 2.5 per lo stato fondamentale, e di 5 per lo stato eccitato. La posizione dell'esploratore continua a oscillare attorno a questo valore medio, esplorando la distribuzione di probabilità. In seguito all'equilibrazione, possiamo fargli ora campionare l'intera funzione d'onda, con un milione di passi Monte Carlo:

```python
fig = plt.figure(figsize=(10,8))
plot(1,'Fondamentale',True)
plot(2,'Eccitato',True)
plt.show()
```

Ho scelto per ciascun blocco una dimensione di 5000 passi. Come si vedrà successivamente, la dimensione di un blocco non deve essere troppo piccola perché altrimenti blocchi successivi risulterebbero correlati tra loro. Non deve inoltre essere eccessivamente grande, perché altrimenti avremmo un campione troppo ridotto di dati sperimentali. L'ottimo è dunque una via di mezzo tra queste due necessità.

Effettuando una media a blocchi, otteniamo per $\langle r\rangle$ la seguente stima, che migliora all'aumentare del numero di blocchi (esperimenti):

```python
import matplotlib.pyplot as plt
import numpy as np

r100, err100 = np.loadtxt('risultati/rMedioFondamentaleUniforme.txt',unpack=True,usecols=(0,1))
r100G, err100G = np.loadtxt('risultati/rMedioFondamentaleGaussiano.txt',unpack=True,usecols=(0,1))
r210, err210 = np.loadtxt('risultati/rMedioEccitatoUniforme.txt',unpack=True,usecols=(0,1))
r210G, err210G = np.loadtxt('risultati/rMedioEccitatoGaussiano.txt',unpack=True,usecols=(0,1))

x = np.arange(r100.size)

plt.figure(figsize=(14,5))
plt.errorbar(x,r100,yerr=err100,alpha=0.8, label="Passo uniforme", color="orange")
plt.errorbar(x,r100G,yerr=err100G,alpha=0.8, label="Passo Gaussiano",color='purple')
plt.title("<r> | Stato Fondamentale", fontsize=17)
plt.xlabel('Blocchi')
plt.ylabel('<r>')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(14,5))
plt.errorbar(x,r210,yerr=err210,alpha=0.8, label="Passo uniforme", color="orange")
plt.errorbar(x,r210G,yerr=err210G,alpha=0.8, label="Passo Gaussiano",color='purple')
plt.title("<r> | Stato Eccitato", fontsize=17)
plt.xlabel('Blocchi')
plt.ylabel('<r>')
plt.legend()
plt.grid(True)
plt.show()
```

Non noto una differenza sostanziale tra il campionamento con passo uniforme o gaussiano, i risultati sono compatibili con il valor medio teorico nel limite di $\sigma$.
